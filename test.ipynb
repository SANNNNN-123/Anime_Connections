````python
import requests
import json
import os
from datetime import datetime

def get_trivia_questions(amount=10, difficulty='easy'):
    """
    Fetch trivia questions from Open Trivia Database API
    
    :param amount: Number of questions to retrieve (default 10)
    :param difficulty: Question difficulty ('easy', 'medium', 'hard') (default 'easy')
    :return: List of trivia questions or None if error occurs
    """
    # Validate difficulty
    valid_difficulties = ['easy', 'medium', 'hard']
    if difficulty.lower() not in valid_difficulties:
        print(f"Invalid difficulty. Choose from {valid_difficulties}")
        return None

    # Construct API URL
    url = f"https://opentdb.com/api.php?amount={amount}&category=31&difficulty={difficulty}&type=multiple"
    
    try:
        # Send GET request to the API
        response = requests.get(url)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the JSON response
            data = response.json()
            return data
        else:
            print(f"Error: Received status code {response.status_code}")
            return None
    
    except requests.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

def save_trivia_questions(data, filename=None):
    """
    Save trivia questions to a JSON file
    
    :param data: JSON data to save
    :param filename: Custom filename (optional)
    :return: Path to saved file
    """
    # Create 'trivia_data' directory if it doesn't exist
    os.makedirs('trivia_data', exist_ok=True)
    
    # If no filename provided, generate a timestamped filename
    if filename is None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"trivia_data/trivia_questions_{timestamp}.json"
    else:
        # Ensure .json extension
        if not filename.endswith('.json'):
            filename += '.json'
        filename = f"trivia_data/{filename}"
    
    # Save the entire API response to the file
    try:
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        
        print(f"Trivia questions saved to {filename}")
        return filename
    except Exception as e:
        print(f"Error saving file: {e}")
        return None

#maximum is 50
trivia_data = get_trivia_questions(amount=38, difficulty='medium')
if trivia_data:
    # Save the data with a custom filename
    save_trivia_questions(trivia_data, 'anime_medium_questions')
    
    # Print out the questions
    for question in trivia_data['results']:
        print("Question:", question['question'])
        print("Correct Answer:", question['correct_answer'])
        print("---")
````

````python
import google.generativeai as genai
import json
import re
import time
from dotenv import load_dotenv
from prompts import easy_puzzle_prompt,medium_puzzle_prompt
import os

load_dotenv()

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-2.0-flash-exp')

def extract_json_from_text(text):
    """Extracts JSON from text using regex."""
    try:
       json_match = re.search(r'```json\n(.*)\n```', text, re.DOTALL)
       if json_match:
            json_str = json_match.group(1)
       else:
           json_str = text
       
       json_str = json_str.strip()
       return json.loads(json_str)
    except json.JSONDecodeError:
        print("Couldn't parse JSON. Attempting to clean and parse.")
        try:
            # More aggressive cleaning
            json_str = re.sub(r'\s+', ' ', json_str)
            return json.loads(json_str)
        except Exception as e:
            print(f"Final JSON parsing error: {e}")
            return None

def is_valid_puzzle(data):
    """Checks if the generated data is valid for our puzzle format."""
    if not isinstance(data, list):
      return False
    for item in data:
      if (not isinstance(item, dict) or
          "level" not in item or
          "category" not in item or
          "words" not in item or
          not isinstance(item["words"], list) or
          len(item["words"]) != 4):
          return False
    return True


def generate_puzzle_data(level, num_puzzles=2, max_tries=5):
    """Generates a set of connection puzzles using Google AI API."""
    prompt = ""
    if level == "easy":
        prompt = easy_puzzle_prompt
    elif level == "medium":
        num_puzzles= 1
        prompt = medium_puzzle_prompt

    tries = 0
    while tries < max_tries:
        try:
          response = model.generate_content(prompt)
          response_text = response.text
          json_data = extract_json_from_text(response_text)

          if json_data is None:
                print(f"Attempt {tries+1}: Failed to extract JSON")
                tries += 1
                continue

          if is_valid_puzzle(json_data):
                return json_data
          else:
                print(f"Attempt {tries+1}: Invalid puzzle format")
        except Exception as e:
            print(f"Attempt {tries+1} error: {e}")
        tries += 1
    print(f"Failed to generate valid data after {max_tries} attempts for {level} level.")
    return []


def generate_level_data(level, num_categories):
    """Generates a specified number of puzzle sets for a given level.
    Returns a single list of dictionary objects.
    """
    all_data = []
    categories_generated = 0
    while categories_generated < num_categories:
        data_chunk = generate_puzzle_data(level, num_puzzles=1)
        if data_chunk:
            all_data.extend(data_chunk)
            categories_generated +=1
            time.sleep(1)
        else:
          print(f"Failed to generate categories. Retrying for {level} level.")

    return all_data

def generate_and_display(level, num_puzzles):
    """Generates and prints puzzle data for a given level."""
    data = generate_puzzle_data(level, num_puzzles)
    if data:
        print("Generated Data:\n")
        print(json.dumps(data, indent=2))
        #save_puzzle_data_to_json(data, level)

    else:
        print(f"Failed to generate data for {level} level.")

def save_puzzle_data_to_json(data, level, filename=None):

    os.makedirs('puzzles', exist_ok=True)

    if filename is None:
        from datetime import datetime
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"puzzles/{level}_puzzles_{timestamp}.json"
    else:
        filename = os.path.join('puzzles', filename if filename.endswith('.json') else f"{filename}.json")

    try:
        with open(filename, 'w') as f:
            json.dump(data, f, indent=2)
        print(f"Puzzle data saved to {filename}")
        return filename
    except Exception as e:
        print(f"Error saving puzzle data: {e}")
        return None


if __name__ == '__main__':
    # print("Generating Easy Level Puzzles:")
    # generate_and_display("easy", 3)
    print("Generating Medium Level Puzzles:")
    generate_and_display("medium", 1)

````

````python
import json
import os
from collections import Counter

def find_duplicate_words(json_file):
    """
    Reads a JSON file of puzzles and checks for duplicate words across puzzles.

    Args:
        json_file (str): Path to the JSON file containing puzzles

    Returns:
        dict: A dictionary containing information about duplicate words
    """
    try:
        # Read the JSON file
        with open(json_file, 'r') as f:
            puzzles = json.load(f)
        
        # Collect all words from all puzzles
        all_words = []
        for puzzle in puzzles:
            all_words.extend(puzzle.get('words', []))
        
        # Count the occurrences of each word
        word_counts = Counter(all_words)
        
        # Identify duplicates (words that appear more than once)
        duplicates = {word: count for word, count in word_counts.items() if count > 1}
        
        # Detailed duplicate analysis
        duplicate_details = {}
        if duplicates:
            for word, count in duplicates.items():
                # Find which puzzles contain the duplicate word
                duplicate_puzzles = [
                    {
                        'category': puzzle['category'], 
                        'level': puzzle['level']
                    } 
                    for puzzle in puzzles 
                    if word in puzzle.get('words', [])
                ]
                
                duplicate_details[word] = {
                    'count': count,
                    'puzzles': duplicate_puzzles
                }
        
        return {
            'has_duplicates': bool(duplicates),
            'total_words': len(all_words),
            'unique_words': len(set(all_words)),
            'duplicate_words': duplicate_details
        }
    
    except FileNotFoundError:
        return {
            'error': f"File not found: {json_file}",
            'has_duplicates': False
        }
    except json.JSONDecodeError:
        return {
            'error': f"Invalid JSON in file: {json_file}",
            'has_duplicates': False
        }
    except Exception as e:
        return {
            'error': str(e),
            'has_duplicates': False
        }

# Example usage function
def analyze_puzzle_duplicates(json_file):
    """
    Prints a detailed analysis of duplicate words in puzzles.
    
    Args:
        json_file (str): Path to the JSON file containing puzzles
    """
    result = find_duplicate_words(json_file)
    
    print("\n--- Puzzle Duplicate Analysis ---")
    print(f"Total Words: {result.get('total_words', 0)}")
    print(f"Unique Words: {result.get('unique_words', 0)}")
    
    if result.get('error'):
        print(f"Error: {result['error']}")
        return
    
    if result['has_duplicates']:
        print("\nDuplicate Words Found:")
        for word, details in result['duplicate_words'].items():
            print(f"\n- {word} (appears {details['count']} times)")
            print("  Found in puzzles:")
            for puzzle in details['puzzles']:
                print(f"    - Category: {puzzle['category']} (Level: {puzzle['level']})")
    else:
        print("\nNo duplicate words found.")

# Demonstration
analyze_puzzle_duplicates('puzzles/easy_puzzles_20241215_045446.json')
````

````python
import google.generativeai as genai
import json
import re
import time
from dotenv import load_dotenv
from prompts import easy_puzzle_prompt, medium_puzzle_prompt
import os

load_dotenv()

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-2.0-flash-exp')


def extract_json_from_text(text):
    """Extracts JSON from text, with added aggressive parsing techniques for broken json strings."""
    try:
        json_match = re.search(r'```json\n(.*)\n```', text, re.DOTALL)
        if json_match:
           json_str = json_match.group(1).strip()
        else:
            json_str = text.strip()

        json_str = json_str.replace('\n', '').replace('\\','')  

        json_str = re.sub(r'\s*([{,])\s*', r'\1', json_str)
        
        loaded = json.loads(json_str) 
        if not isinstance(loaded, list):
           return [loaded] 

        return loaded 

    except json.JSONDecodeError as e:
        print(f"Error extracting json with more format try: {e} . Returning Non valid format as dict format error")
        return None


def is_valid_puzzle(data, level = None):
    """Checks if the generated data is valid for our puzzle format for given level format"""
    if not isinstance(data, list) or not data : 
      return False

    if level == "easy": 
         for item in data:
            if (not isinstance(item, dict) or
                "level" not in item or
                "category" not in item or
                "words" not in item or
                not isinstance(item["words"], list) or
                len(item["words"]) != 4):
              return False

         return True 


    elif level == "medium":
        for item in data:

              if (not isinstance(item, dict) or
                    "level" not in item or
                  "categories" not in item or
                  not isinstance(item["categories"], dict) or
                 len(item["categories"]) != 4) : # added medium format specific check for categories

                    return False

              for category, words in item["categories"].items() : # specific for all four nested cat members to check validity
                if not isinstance(words,list) or len(words) != 4:
                      return False

        return True  #return that this data matches.
    else :
           return False


def generate_puzzle_data(level, num_puzzles=2, max_tries=5):
    """Generates a set of connection puzzles using Google AI API."""
    prompt = ""
    if level == "easy":
        prompt = easy_puzzle_prompt
    elif level == "medium":
        num_puzzles= 1
        prompt = medium_puzzle_prompt

    tries = 0
    while tries < max_tries:
        try:
          response = model.generate_content(prompt)
          response_text = response.text
          json_data = extract_json_from_text(response_text)

          if json_data is None:
                print(f"Attempt {tries+1}: Failed to extract JSON")
                tries += 1
                continue

          if is_valid_puzzle(json_data, level = level):
                return json_data
          else:
                print(f"Attempt {tries+1}: Invalid puzzle format")
        except Exception as e:
            print(f"Attempt {tries+1} error: {e}")
        tries += 1
    print(f"Failed to generate valid data after {max_tries} attempts for {level} level.")
    return []


def generate_level_data(level, num_categories):
    """Generates a specified number of puzzle sets for a given level.
    Returns a single list of dictionary objects.
    """
    all_data = []
    categories_generated = 0
    while categories_generated < num_categories:
        data_chunk = generate_puzzle_data(level, num_puzzles=1)
        if data_chunk:
            all_data.extend(data_chunk)
            categories_generated +=1
            time.sleep(1)
        else:
          print(f"Failed to generate categories. Retrying for {level} level.")

    return all_data

def generate_and_display(level, num_puzzles):
    """Generates and prints puzzle data for a given level."""
    data = generate_level_data(level, num_puzzles)
    if data:
        print("Generated Data:\n")
        print(json.dumps(data, indent=2))
        #save_puzzle_data_to_json(data, level)

    else:
        print(f"Failed to generate data for {level} level.")

def save_puzzle_data_to_json(data, level, filename=None):

    os.makedirs('puzzles', exist_ok=True)

    if filename is None:
        from datetime import datetime
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"puzzles/{level}_puzzles_{timestamp}.json"
    else:
        filename = os.path.join('puzzles', filename if filename.endswith('.json') else f"{filename}.json")

    try:
        with open(filename, 'w') as f:
            json.dump(data, f, indent=2)
        print(f"Puzzle data saved to {filename}")
        return filename
    except Exception as e:
        print(f"Error saving puzzle data: {e}")
        return None


if __name__ == '__main__':
    # print("Generating Easy Level Puzzles:")
    # generate_and_display("easy", 3)
    print("Generating Medium Level Puzzles:")
    generate_and_display("medium", 1)
````
