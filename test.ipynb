````python
import requests
import json
import os
from datetime import datetime

def get_trivia_questions(amount=10, difficulty='easy'):
    """
    Fetch trivia questions from Open Trivia Database API
    
    :param amount: Number of questions to retrieve (default 10)
    :param difficulty: Question difficulty ('easy', 'medium', 'hard') (default 'easy')
    :return: List of trivia questions or None if error occurs
    """
    # Validate difficulty
    valid_difficulties = ['easy', 'medium', 'hard']
    if difficulty.lower() not in valid_difficulties:
        print(f"Invalid difficulty. Choose from {valid_difficulties}")
        return None

    # Construct API URL
    url = f"https://opentdb.com/api.php?amount={amount}&category=31&difficulty={difficulty}&type=multiple"
    
    try:
        # Send GET request to the API
        response = requests.get(url)
        
        # Check if the request was successful
        if response.status_code == 200:
            # Parse the JSON response
            data = response.json()
            return data
        else:
            print(f"Error: Received status code {response.status_code}")
            return None
    
    except requests.RequestException as e:
        print(f"Error fetching data: {e}")
        return None

def save_trivia_questions(data, filename=None):
    """
    Save trivia questions to a JSON file
    
    :param data: JSON data to save
    :param filename: Custom filename (optional)
    :return: Path to saved file
    """
    # Create 'trivia_data' directory if it doesn't exist
    os.makedirs('trivia_data', exist_ok=True)
    
    # If no filename provided, generate a timestamped filename
    if filename is None:
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"trivia_data/trivia_questions_{timestamp}.json"
    else:
        # Ensure .json extension
        if not filename.endswith('.json'):
            filename += '.json'
        filename = f"trivia_data/{filename}"
    
    # Save the entire API response to the file
    try:
        with open(filename, 'w', encoding='utf-8') as f:
            json.dump(data, f, ensure_ascii=False, indent=4)
        
        print(f"Trivia questions saved to {filename}")
        return filename
    except Exception as e:
        print(f"Error saving file: {e}")
        return None

#maximum is 50
trivia_data = get_trivia_questions(amount=38, difficulty='medium')
if trivia_data:
    # Save the data with a custom filename
    save_trivia_questions(trivia_data, 'anime_medium_questions')
    
    # Print out the questions
    for question in trivia_data['results']:
        print("Question:", question['question'])
        print("Correct Answer:", question['correct_answer'])
        print("---")
````

````python
import google.generativeai as genai
import json
import re
import time
from dotenv import load_dotenv
from prompts import easy_puzzle_prompt,medium_puzzle_prompt
import os

load_dotenv()

GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
genai.configure(api_key=GOOGLE_API_KEY)
model = genai.GenerativeModel('gemini-2.0-flash-exp')

def extract_json_from_text(text):
    """Extracts JSON from text using regex."""
    try:
       json_match = re.search(r'```json\n(.*)\n```', text, re.DOTALL)
       if json_match:
            json_str = json_match.group(1)
       else:
           json_str = text
       
       json_str = json_str.strip()
       return json.loads(json_str)
    except json.JSONDecodeError:
        print("Couldn't parse JSON. Attempting to clean and parse.")
        try:
            # More aggressive cleaning
            json_str = re.sub(r'\s+', ' ', json_str)
            return json.loads(json_str)
        except Exception as e:
            print(f"Final JSON parsing error: {e}")
            return None

def is_valid_puzzle(data):
    """Checks if the generated data is valid for our puzzle format."""
    if not isinstance(data, list):
      return False
    for item in data:
      if (not isinstance(item, dict) or
          "level" not in item or
          "category" not in item or
          "words" not in item or
          not isinstance(item["words"], list) or
          len(item["words"]) != 4):
          return False
    return True


def generate_puzzle_data(level, num_puzzles=2, max_tries=5):
    """Generates a set of connection puzzles using Google AI API."""
    prompt = ""
    if level == "easy":
        prompt = easy_puzzle_prompt
    elif level == "medium":
        num_puzzles= 1
        prompt = medium_puzzle_prompt

    tries = 0
    while tries < max_tries:
        try:
          response = model.generate_content(prompt)
          response_text = response.text
          json_data = extract_json_from_text(response_text)

          if json_data is None:
                print(f"Attempt {tries+1}: Failed to extract JSON")
                tries += 1
                continue

          if is_valid_puzzle(json_data):
                return json_data
          else:
                print(f"Attempt {tries+1}: Invalid puzzle format")
        except Exception as e:
            print(f"Attempt {tries+1} error: {e}")
        tries += 1
    print(f"Failed to generate valid data after {max_tries} attempts for {level} level.")
    return []


def generate_level_data(level, num_categories):
    """Generates a specified number of puzzle sets for a given level.
    Returns a single list of dictionary objects.
    """
    all_data = []
    categories_generated = 0
    while categories_generated < num_categories:
        data_chunk = generate_puzzle_data(level, num_puzzles=1)
        if data_chunk:
            all_data.extend(data_chunk)
            categories_generated +=1
            time.sleep(1)
        else:
          print(f"Failed to generate categories. Retrying for {level} level.")

    return all_data

def generate_and_display(level, num_puzzles):
    """Generates and prints puzzle data for a given level."""
    data = generate_puzzle_data(level, num_puzzles)
    if data:
        print("Generated Data:\n")
        print(json.dumps(data, indent=2))
        #save_puzzle_data_to_json(data, level)

    else:
        print(f"Failed to generate data for {level} level.")

def save_puzzle_data_to_json(data, level, filename=None):

    os.makedirs('puzzles', exist_ok=True)

    if filename is None:
        from datetime import datetime
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        filename = f"puzzles/{level}_puzzles_{timestamp}.json"
    else:
        filename = os.path.join('puzzles', filename if filename.endswith('.json') else f"{filename}.json")

    try:
        with open(filename, 'w') as f:
            json.dump(data, f, indent=2)
        print(f"Puzzle data saved to {filename}")
        return filename
    except Exception as e:
        print(f"Error saving puzzle data: {e}")
        return None


if __name__ == '__main__':
    # print("Generating Easy Level Puzzles:")
    # generate_and_display("easy", 3)
    print("Generating Medium Level Puzzles:")
    generate_and_display("medium", 1)

````

````python
import json
import os
from collections import Counter

def find_duplicate_words(json_file):
    """
    Reads a JSON file of puzzles and checks for duplicate words across puzzles.

    Args:
        json_file (str): Path to the JSON file containing puzzles

    Returns:
        dict: A dictionary containing information about duplicate words
    """
    try:
        # Read the JSON file
        with open(json_file, 'r') as f:
            puzzles = json.load(f)
        
        # Collect all words from all puzzles
        all_words = []
        for puzzle in puzzles:
            all_words.extend(puzzle.get('words', []))
        
        # Count the occurrences of each word
        word_counts = Counter(all_words)
        
        # Identify duplicates (words that appear more than once)
        duplicates = {word: count for word, count in word_counts.items() if count > 1}
        
        # Detailed duplicate analysis
        duplicate_details = {}
        if duplicates:
            for word, count in duplicates.items():
                # Find which puzzles contain the duplicate word
                duplicate_puzzles = [
                    {
                        'category': puzzle['category'], 
                        'level': puzzle['level']
                    } 
                    for puzzle in puzzles 
                    if word in puzzle.get('words', [])
                ]
                
                duplicate_details[word] = {
                    'count': count,
                    'puzzles': duplicate_puzzles
                }
        
        return {
            'has_duplicates': bool(duplicates),
            'total_words': len(all_words),
            'unique_words': len(set(all_words)),
            'duplicate_words': duplicate_details
        }
    
    except FileNotFoundError:
        return {
            'error': f"File not found: {json_file}",
            'has_duplicates': False
        }
    except json.JSONDecodeError:
        return {
            'error': f"Invalid JSON in file: {json_file}",
            'has_duplicates': False
        }
    except Exception as e:
        return {
            'error': str(e),
            'has_duplicates': False
        }

# Example usage function
def analyze_puzzle_duplicates(json_file):
    """
    Prints a detailed analysis of duplicate words in puzzles.
    
    Args:
        json_file (str): Path to the JSON file containing puzzles
    """
    result = find_duplicate_words(json_file)
    
    print("\n--- Puzzle Duplicate Analysis ---")
    print(f"Total Words: {result.get('total_words', 0)}")
    print(f"Unique Words: {result.get('unique_words', 0)}")
    
    if result.get('error'):
        print(f"Error: {result['error']}")
        return
    
    if result['has_duplicates']:
        print("\nDuplicate Words Found:")
        for word, details in result['duplicate_words'].items():
            print(f"\n- {word} (appears {details['count']} times)")
            print("  Found in puzzles:")
            for puzzle in details['puzzles']:
                print(f"    - Category: {puzzle['category']} (Level: {puzzle['level']})")
    else:
        print("\nNo duplicate words found.")

# Demonstration
analyze_puzzle_duplicates('puzzles/easy_puzzles_20241215_045446.json')
````

````python
import random
import google.generativeai as genai
import os
import re

# Set your Google Gemini API key
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")
if not GOOGLE_API_KEY:
    print("Error: Google Gemini API key not found. Please set your GOOGLE_API_KEY environment variable.")
else:
    genai.configure(api_key=GOOGLE_API_KEY)

generation_config = {
    "temperature": 1,
    "top_p": 1,
    "top_k": 10,
    "max_output_tokens": 2048
}

model = genai.GenerativeModel(
    model_name="gemini-2.0-flash-exp",  # Or "gemini-pro", "gemini-2.0-flash-exp" depending on your need.
    generation_config=generation_config,
)

def proposal_agent(num_characters=8):
    """
    Generates a detailed categorization of 8 random One Piece characters using Gemini.

    Args:
        num_characters: The number of unique characters to select. Default is 8.

    Returns:
         str: A String containing the categories and associated details, or None if an error occurs.
    """

    if not GOOGLE_API_KEY:
        return None

    try:
        # 1. Generate random characters using Gemini
        character_prompt = f"""
            Generate {num_characters} random characters from One Piece anime, 
            separated by commas, do not include any other text. For example: 
             Luffy, Zoro, Nami, Usopp, Sanji, Chopper, Robin, Franky
            """

        response = model.generate_content(character_prompt)

        characters_str = response.text.strip()
        selected_characters = [char.strip() for char in characters_str.split(",")]

        if len(selected_characters) != num_characters:
            print(f"Error: Gemini returned {len(selected_characters)} characters, expected {num_characters}.")
            return None

        # 2. Generate categories using Gemini
        category_prompt = f"""
           Given the following One Piece characters: {', '.join(selected_characters)}, 
           perform the following:

            1. Analyze the characters for shared traits, focusing on:
              - Character backgrounds
              - Story roles
              - Abilities/powers
              - Narrative themes
              - Personal relationships
              - Timeline appearances

            2. Create FOUR groups of TWO characters each based on the strongest thematic or characteristic similarities

            3. For each group provide the following details:
                - Provide a clear theme/title (2-3 words)
                - List the 2 characters
                - Assign a similarity score (1-10)

            Only show output of group and characters.
            Example output:
            
            Output format:

            ## [Group 1 Theme]
            Characters: [Name 1], [Name 2]
            Similarity: [X/10]

            ## [Group 2 Theme]
            Characters: [Name 1], [Name 2]
            Similarity: [X/10]

            ## [Group 3 Theme]
            Characters: [Name 1], [Name 2]
            Similarity: [X/10]

            ## [Group 4 Theme]
            Characters: [Name 1], [Name 2]
            Similarity: [X/10]
            """

        response = model.generate_content(category_prompt)
        categories_str = response.text.strip()
        return selected_characters, categories_str

    except Exception as e:
        print(f"An error occurred: {e}")
        return None
    
    
def grouping_agent(characters, initial_categories_str):
    """
    Expands each two-character group into a four-character group while maintaining thematic consistency.

    Args:
        characters (list): List of the 8 characters
        initial_categories_str (str): String containing the initial categories and characters

    Returns:
        str: A String containing the expanded categories and characters in markdown format, or None if an error occurs.
    """
    if not characters or not initial_categories_str:
      return None
    
    try:
         # 3. Expand categories using Gemini
         grouping_prompt = f"""
            Given the following One Piece characters: {', '.join(characters)}, 
           and the following groups:
            {initial_categories_str}
           
            Your task is to expand each two-character group into a four-character group while maintaining thematic consistency. 
               Requirements:
                    - Add EXACTLY 2 more characters to each group
                    - New characters must:
                        - Clearly match the group's theme
                        - Do Not repeat characters in multiple groups
                        - Be canonical anime characters
                        - Have clear justification for inclusion
                    - Each group must have exactly 4 unique characters
                    - Maintain thematic consistency within each group

           Only output the group and characters, do not put description

           Output format:

           ## [Group 1 Theme]
           Characters: [Name 1], [Name 2], [Name 3], [Name 4]

           ## [Group 2 Theme]
           Characters: [Name 1], [Name 2], [Name 3], [Name 4]

           ## [Group 3 Theme]
           Characters: [Name 1], [Name 2], [Name 3], [Name 4]

           ## [Group 4 Theme]
           Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            """

         response = model.generate_content(grouping_prompt)
         expanded_categories_str = response.text.strip()
         return expanded_categories_str
    except Exception as e:
        print(f"An error occurred: {e}")
        return None

def editing_agent(grouped_output):
    """
    Scans the grouped output for duplicate characters, replaces them, and
    adjusts themes as needed.

    Args:
        grouped_output (str): Output from the grouping_agent in markdown format.

    Returns:
        str: The edited grouped output string, or None if an error occurs.
    """

    if not grouped_output:
        return None

    try:
        # 1. Parse the output
        groups = {}
        lines = grouped_output.split("\n")
        current_theme = None
        for line in lines:
            if line.startswith("## "):
                current_theme = line[3:].strip()
                groups[current_theme] = []
            elif line.startswith("Characters: "):
                characters = [char.strip() for char in line[11:].split(",")]
                if current_theme:
                    groups[current_theme] = characters

        # 2. Check for duplicate characters
        all_chars = []
        for theme, chars in groups.items():
            all_chars.extend(chars)
        
        duplicates = set([char for char in all_chars if all_chars.count(char) > 1])


        # 3. Handle duplicates (if any)
        if duplicates:
            editing_prompt = f"""
                Given the following groups and duplicate characters:
                Groups: {groups}
                Duplicate characters: {list(duplicates)}
    
                Your task is to modify groups. To do this, you must:
                   1. Scan for duplicate characters across all groups
                   2. Identify thematically appropriate replacements, do not repeat
                    3. Maintain group theme integrity
                   4.Replace the duplicate character.
                Only output the group and characters, do not put description. Use following format:

                 Output format:

                    ## [Group 1 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            
                    ## [Group 2 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            
                    ## [Group 3 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            
                    ## [Group 4 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
                """
            response = model.generate_content(editing_prompt)
            edited_output = response.text.strip()
            return edited_output
        
       # 4. Check if themes make sense and modify them.
        theme_check_prompt = f"""
           Given the following groups:
            {groups}

            Your task is to:
                - Check if the group theme is appropriate and make it simple words, modify if not
            Only output the group and characters, do not put description. Use following format:

                 Output format:

                    ## [Group 1 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            
                    ## [Group 2 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            
                    ## [Group 3 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
            
                    ## [Group 4 Theme]
                    Characters: [Name 1], [Name 2], [Name 3], [Name 4]
        """
        response = model.generate_content(theme_check_prompt)
        edited_output = response.text.strip()
        return edited_output

    except Exception as e:
      print(f"An error occurred during editing: {e}")
      return None


characters, initial_categories = proposal_agent()
print(initial_categories)
print("========================================================================================================")
if characters and initial_categories:
    expanded_categories = grouping_agent(characters, initial_categories)
    if expanded_categories:
        print(expanded_categories)
        print("========================================================================================================")
        edited_categories = editing_agent(expanded_categories)
        if edited_categories:
           print(edited_categories)
````
